```python
import plotly.graph_objects as go
import numpy as np
```

```python
step  = lambda x : 0 if x < 0 else 1
sigma = lambda x : 1 / (1 + np.exp(-x))
ReLU  = lambda x : x * (x > 0)

step  = np.vectorize(step)
sigma = np.vectorize(sigma)
ReLU  = np.vectorize(ReLU)
```

```python
x = np.linspace(-5, 5, 10000)
```

```python
y = step(x)

fig = go.Figure(
    go.Scatter(
        x=x,
        y=y,
        mode='lines',
        line_shape='hv',  
        marker=dict(size=8)
    )
)

fig.update_layout(
    title={
        'text': 'Step Function',
        'x': 0.5,
        'font': dict(size=20)
    },
    template="plotly_white",
    autosize=False,
    width=500, height=500,
    # margin=dict(l=30, r=30, t=30, b=30),
)
fig.update_xaxes(tickfont_size=15)
fig.update_yaxes(tickfont_size=15)

fig.show()
```

```python
y = sigma(x)

fig = go.Figure(
    go.Scatter(
        x=x,
        y=y,
        mode='lines',
        line_shape='hv',  
        marker=dict(size=8)
    )
)

fig.update_layout(
    title={
        'text': 'Sigmoid Function',
        'x': 0.5,
        'font': dict(size=20)
    },
    template="plotly_white",
    autosize=False,
    width=500, height=500,
    # margin=dict(l=30, r=30, t=30, b=30),
)
fig.update_xaxes(tickfont_size=15)
fig.update_yaxes(tickfont_size=15)

fig.show()
```

```python
y = np.tanh(x)

fig = go.Figure(
    go.Scatter(
        x=x,
        y=y,
        mode='lines',
        line_shape='hv',  
        marker=dict(size=8)
    )
)

fig.update_layout(
    title={
        'text': 'Hyperbolic Tangent Function',
        'x': 0.5,
        'font': dict(size=20)
    },
    template="plotly_white",
    autosize=False,
    width=500, height=500,
    # margin=dict(l=30, r=30, t=30, b=30),
)
fig.update_xaxes(tickfont_size=15)
fig.update_yaxes(tickfont_size=15)

fig.show()
```

```python
y = ReLU(x)

fig = go.Figure(
    go.Scatter(
        x=x,
        y=y,
        mode='lines',
        line_shape='hv',  
        marker=dict(size=8)
    )
)

fig.update_layout(
    title={
        'text': 'Rectified Linear Unit',
        'x': 0.5,
        'font': dict(size=20)
    },
    template="plotly_white",
    autosize=False,
    width=500, height=500,
    # margin=dict(l=30, r=30, t=30, b=30),
)
fig.update_xaxes(tickfont_size=15)
fig.update_yaxes(tickfont_size=15)

fig.show()
```
